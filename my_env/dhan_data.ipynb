```python
# functons
from datetime import datetime

def convert_epoch_time_to_real_time(epoch_time):
    # epoch_time+=315528518.0
    date_time = datetime.datetime.fromtimestamp( epoch_time )
    return date_time.strftime('%Y-%m-%d %H:%M:%S')


def round_to_nearest_5_paisa(value):
    return round(value * 20) / 20


def aggregate_to_5min(data_dict):
    # Convert start times to datetime format
    start_times = [datetime.fromtimestamp(t) for t in data_dict["start_Time"]]

    # Initialize empty lists for aggregated OHLC data
    open_5min = []  
    high_5min = []
    low_5min = []
    close_5min = []
    volume_5min = []
    start_times_5min = []

    # Initialize variables for aggregation
    start_index = 0
    end_index = 4

    while end_index < len(start_times):
        # Aggregate OHLC data for each 5-minute interval
        open_5min.append(data_dict["open"][start_index])
        high_5min.append(max(data_dict["high"][start_index:end_index+1]))
        low_5min.append(min(data_dict["low"][start_index:end_index+1]))
        close_5min.append(data_dict["close"][end_index])
        volume_5min.append(sum(data_dict["volume"][start_index:end_index+1]))
        start_times_5min.append(start_times[start_index])

        # Move to the next 5-minute interval
        start_index += 5
        end_index += 5

    # Add the current incomplete 5-minute interval if available
    if start_index < len(start_times):
        open_5min.append(data_dict["open"][start_index])
        high_5min.append(max(data_dict["high"][start_index:]))
        low_5min.append(min(data_dict["low"][start_index:]))
        close_5min.append(data_dict["close"][-1])  # Assuming last close is the current close
        volume_5min.append(sum(data_dict["volume"][start_index:]))
        start_times_5min.append(start_times[start_index])

    # Create dictionary to hold aggregated data
    aggregated_data = {
        "open": open_5min,
        "high": high_5min,
        "low": low_5min,
        "close": close_5min,
        "volume": volume_5min,
        "start_Time": [int(t.timestamp()) for t in start_times_5min]
    }

    return aggregated_data


def aggregate_to_15min(data_dict):
    # Convert start times to datetime format
    start_times = [datetime.fromtimestamp(t) for t in data_dict["start_Time"]]

    # Initialize empty lists for aggregated OHLC data
    open_15min = []  
    high_15min = []
    low_15min = []
    close_15min = []
    volume_15min = []
    start_times_15min = []

    # Initialize variables for aggregation
    interval = 15  # 15-minute interval
    num_bars = interval // 1  # Since each bar is 1 minute
    start_index = 0
    end_index = num_bars - 1

    while end_index < len(start_times):
        # Open is the first open price in the 15-minute block
        open_15min.append(data_dict["open"][start_index])

        # High is the highest price in the 15-minute block
        high_15min.append(max(data_dict["high"][start_index:end_index + 1]))

        # Low is the lowest price in the 15-minute block
        low_15min.append(min(data_dict["low"][start_index:end_index + 1]))

        # Close is the last close price in the 15-minute block
        close_15min.append(data_dict["close"][end_index])

        # Volume is the sum of the volumes in the 15-minute block
        volume_15min.append(sum(data_dict["volume"][start_index:end_index + 1]))

        # The start time of the 15-minute block is the start time of the first minute
        start_times_15min.append(start_times[start_index])

        # Move to the next 15-minute block
        start_index += num_bars
        end_index += num_bars

    # Add the current incomplete 15-minute interval if available
    if start_index < len(start_times):
        open_15min.append(data_dict["open"][start_index])
        high_15min.append(max(data_dict["high"][start_index:]))
        low_15min.append(min(data_dict["low"][start_index:]))
        close_15min.append(data_dict["close"][-1])  # Assuming last close is the current close
        volume_15min.append(sum(data_dict["volume"][start_index:]))
        start_times_15min.append(start_times[start_index])

    # Create dictionary to hold aggregated data
    aggregated_data = {
        "open": open_15min,
        "high": high_15min,
        "low": low_15min,
        "close": close_15min,
        "volume": volume_15min,
        "start_Time": [int(t.timestamp()) for t in start_times_15min]
    }

    return aggregated_data


```
