```python
# functons
from datetime import datetime

def convert_epoch_time_to_real_time(epoch_time):
    # epoch_time+=315528518.0
    date_time = datetime.datetime.fromtimestamp( epoch_time )
    return date_time.strftime('%Y-%m-%d %H:%M:%S')


def round_to_nearest_5_paisa(value):
    return round(value * 20) / 20


def aggregate_to_5min(data_dict):
    # Convert start times to datetime format
    start_times = [datetime.fromtimestamp(t) for t in data_dict["start_Time"]]

    # Initialize empty lists for aggregated OHLC data
    open_5min = []  
    high_5min = []
    low_5min = []
    close_5min = []
    volume_5min = []
    start_times_5min = []

    # Initialize variables for aggregation
    start_index = 0
    end_index = 4

    while end_index < len(start_times):
        # Aggregate OHLC data for each 5-minute interval
        open_5min.append(data_dict["open"][start_index])
        high_5min.append(max(data_dict["high"][start_index:end_index+1]))
        low_5min.append(min(data_dict["low"][start_index:end_index+1]))
        close_5min.append(data_dict["close"][end_index])
        volume_5min.append(sum(data_dict["volume"][start_index:end_index+1]))
        start_times_5min.append(start_times[start_index])

        # Move to the next 5-minute interval
        start_index += 5
        end_index += 5

    # Add the current incomplete 5-minute interval if available
    if start_index < len(start_times):
        open_5min.append(data_dict["open"][start_index])
        high_5min.append(max(data_dict["high"][start_index:]))
        low_5min.append(min(data_dict["low"][start_index:]))
        close_5min.append(data_dict["close"][-1])  # Assuming last close is the current close
        volume_5min.append(sum(data_dict["volume"][start_index:]))
        start_times_5min.append(start_times[start_index])

    # Create dictionary to hold aggregated data
    aggregated_data = {
        "open": open_5min,
        "high": high_5min,
        "low": low_5min,
        "close": close_5min,
        "volume": volume_5min,
        "start_Time": [int(t.timestamp()) for t in start_times_5min]
    }

    return aggregated_data


def aggregate_to_15min(data_dict):
    # Convert start times to datetime format
    start_times = [datetime.fromtimestamp(t) for t in data_dict["start_Time"]]

    # Initialize empty lists for aggregated OHLC data
    open_15min = []  
    high_15min = []
    low_15min = []
    close_15min = []
    volume_15min = []
    start_times_15min = []

    # Initialize variables for aggregation
    interval = 15  # 15-minute interval
    num_bars = interval // 1  # Since each bar is 1 minute
    start_index = 0
    end_index = num_bars - 1

    while end_index < len(start_times):
        # Open is the first open price in the 15-minute block
        open_15min.append(data_dict["open"][start_index])

        # High is the highest price in the 15-minute block
        high_15min.append(max(data_dict["high"][start_index:end_index + 1]))

        # Low is the lowest price in the 15-minute block
        low_15min.append(min(data_dict["low"][start_index:end_index + 1]))

        # Close is the last close price in the 15-minute block
        close_15min.append(data_dict["close"][end_index])

        # Volume is the sum of the volumes in the 15-minute block
        volume_15min.append(sum(data_dict["volume"][start_index:end_index + 1]))

        # The start time of the 15-minute block is the start time of the first minute
        start_times_15min.append(start_times[start_index])

        # Move to the next 15-minute block
        start_index += num_bars
        end_index += num_bars

    # Add the current incomplete 15-minute interval if available
    if start_index < len(start_times):
        open_15min.append(data_dict["open"][start_index])
        high_15min.append(max(data_dict["high"][start_index:]))
        low_15min.append(min(data_dict["low"][start_index:]))
        close_15min.append(data_dict["close"][-1])  # Assuming last close is the current close
        volume_15min.append(sum(data_dict["volume"][start_index:]))
        start_times_15min.append(start_times[start_index])

    # Create dictionary to hold aggregated data
    aggregated_data = {
        "open": open_15min,
        "high": high_15min,
        "low": low_15min,
        "close": close_15min,
        "volume": volume_15min,
        "start_Time": [int(t.timestamp()) for t in start_times_15min]
    }

    return aggregated_data


```

```python
def Stochastic(df, window, smooth_window):

    stochastic = pd.DataFrame()

    stochastic['%K'] = ((df['close'] - df['low'].rolling(window).min()) \

    / (df['high'].rolling(window).max() - df['low'].rolling(window).min())) * 100

    stochastic['%D'] = stochastic['%K'].rolling(smooth_window).mean()

    stochastic['%SD'] = stochastic['%D'].rolling(smooth_window).mean()

    stochastic['UL'] = 80

    stochastic['DL'] = 20

    return stochastic



def stochastic_rsi(df, window=14, smooth_window=3):
    # Calculate RSI
    delta = df['close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))

    # Calculate Stochastic RSI
    min_rsi = rsi.rolling(window=window).min()
    max_rsi = rsi.rolling(window=window).max()

    stoch_rsi = (rsi - min_rsi) / (max_rsi - min_rsi)

    # %K is the Stochastic RSI
    k_percent = stoch_rsi * 100

    # %D is the moving average of %K
    d_percent = k_percent.rolling(window=smooth_window).mean()

    # Add to DataFrame
    df['%K'] = k_percent
    df['%D'] = d_percent

    return df
```

```python
import numpy as np

# Function to calculate RSI (like TradingView)
def RSI(series, period=14):
    delta = series.diff().dropna()
    ups = delta * 0
    downs = ups.copy()
    ups[delta > 0] = delta[delta > 0]
    downs[delta < 0] = -delta[delta < 0]
    ups[ups.index[period-1]] = np.mean( ups[:period] ) # first value is sum of avg gains
    ups = ups.drop(ups.index[:(period-1)])
    downs[downs.index[period-1]] = np.mean( downs[:period] ) # first value is sum of avg losses
    downs = downs.drop(downs.index[:(period-1)])
    rs = ups.ewm(com=period-1,min_periods=0,adjust=False,ignore_na=False).mean() / \
         downs.ewm(com=period-1,min_periods=0,adjust=False,ignore_na=False).mean()
    return 100 - 100 / (1 + rs)

# Function to calculate Stochastic RSI (like TradingView)
def StochRSI(series, period=14, smoothK=3, smoothD=3):
    rsi = RSI(series, period)
    rsi_min = rsi.rolling(window=period).min()
    rsi_max = rsi.rolling(window=period).max()
    stoch_rsi = 100 * (rsi - rsi_min) / (rsi_max - rsi_min)

    k = stoch_rsi.rolling(window=smoothK).mean()  # %K line
    d = k.rolling(window=smoothD).mean()  # %D line

    return k, d
```

```python
import numpy as np

def StochRSI(series, period=14, smoothK=3, smoothD=3):
    # Calculate RSI 
    delta = series.diff().dropna()
    ups = delta * 0
    downs = ups.copy()
    ups[delta > 0] = delta[delta > 0]
    downs[delta < 0] = -delta[delta < 0]
    ups[ups.index[period-1]] = np.mean( ups[:period] ) #first value is sum of avg gains
    ups = ups.drop(ups.index[:(period-1)])
    downs[downs.index[period-1]] = np.mean( downs[:period] ) #first value is sum of avg losses
    downs = downs.drop(downs.index[:(period-1)])
    rs = ups.ewm(com=period-1,min_periods=0,adjust=False,ignore_na=False).mean() / \
         downs.ewm(com=period-1,min_periods=0,adjust=False,ignore_na=False).mean() 
    rsi = 100 - 100 / (1 + rs)

    # Calculate StochRSI 
    stochrsi  = (rsi - rsi.rolling(period).min()) / (rsi.rolling(period).max() - rsi.rolling(period).min())
    stochrsi_K = stochrsi.rolling(smoothK).mean()
    stochrsi_D = stochrsi_K.rolling(smoothD).mean()

    return stochrsi, stochrsi_K, stochrsi_D
```
